### This notebook shows the distribution function evolution during the bump-on-tail instability. 
#### The initial conditions are a background plasma in the presence of a beam at v=3.0. We give this a small perturbation and then let the Vlasov-Poisson system evolve

```python
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
import yaml
from jax import config
config.update("jax_enable_x64", True)
from adept import ergoExo
```

### This is how you can run the code. It outputs 3 objects, the direct simulation output, the post-processed output (which is probably what you want), and the mlflow run id. 

```python
with open("configs/bump-on-tail.yaml", 'r') as stream:
    cfg = yaml.safe_load(stream)

exo = ergoExo()
modules = exo.setup(cfg=cfg)
sim_out, post_processed_output, mlflow_run_id = exo(modules)
```

### We can visualize the distribution function evolution using `xarray`'s built-in functionality

```python
f = post_processed_output["dists"]
```

```python
np.abs(f["electron"][::7, :, 2048:].T).plot(col="t", col_wrap=4)
plt.savefig("./bot.png", bbox_inches="tight")
```
